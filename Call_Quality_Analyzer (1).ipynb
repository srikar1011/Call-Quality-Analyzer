```python

!apt-get update -qq
!apt-get install -y -qq ffmpeg
!pip install -q yt-dlp openai-whisper transformers


!yt-dlp -x --audio-format mp3 -o "/content/call_audio.%(ext)s" "https://www.youtube.com/watch?v=4ostqJD3Psc"
!ls -lh /content/call_audio.*


!ffmpeg -y -i /content/call_audio.mp3 -ar 16000 -ac 1 /content/call_audio_16k.wav
!ls -lh /content/call_audio_16k.wav


import whisper
model = whisper.load_model("tiny.en")
result = model.transcribe("/content/call_audio_16k.wav", language="en")
print("==== Transcript ====\n")
print(result['text'])


lines = result['text'].split('. ')
speaker_1_words, speaker_2_words = 0, 0

for i, line in enumerate(lines):
    word_count = len(line.split())
    if i % 2 == 0:
        speaker_1_words += word_count
    else:
        speaker_2_words += word_count

total_words = speaker_1_words + speaker_2_words
s1_ratio = round((speaker_1_words / total_words) * 100, 2)
s2_ratio = round((speaker_2_words / total_words) * 100, 2)

print("\n=== Talk-time ratio ===")
print(f"Speaker 1: {s1_ratio}%")
print(f"Speaker 2: {s2_ratio}%")


questions = result['text'].count('?')
print("\n=== Number of Questions ===")
print("Questions asked:", questions)


longest_monologue = max(len(line.split()) for line in lines)
print("\n=== Longest Monologue ===")
print(f"{longest_monologue} words (approx)")


from transformers import pipeline
sentiment_model = pipeline("sentiment-analysis")
sentiment = sentiment_model(result['text'][:512])
print("\n=== Call Sentiment ===")
print(sentiment)


if questions > 5:
    insight = "Customer asked many questions → clarify product details."
elif s1_ratio > 70:
    insight = "Sales rep dominated the conversation → let customer talk more."
else:
    insight = "Balanced conversation → good engagement."

print("\n=== Actionable Insight ===")
print(insight)


print("\n=== Role Prediction ===")
if s1_ratio > s2_ratio:
    print("Speaker 1 is likely the Sales Rep")
else:
    print("Speaker 2 is likely the Sales Rep")
```
